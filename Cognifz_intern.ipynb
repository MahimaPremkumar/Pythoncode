```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import unidecode
import re
#Read the given dataset using "read"
df=pd.read_csv("E:/Python Practice/Data_set(Cognifyz).csv")
```

```python
#"Head" to view 1st 5 rows of the table
df.tail(50)
```

```python
#Cleaning data with special characters(Few unwanted special characters are present in the dataset)
def clean_text(text):
    if isinstance(text, str):
        text = unidecode.unidecode(text)  
        text = re.sub(r'[^\w\s,]', '', text)  
        text = text.strip()  
    return text

string_columns = df.select_dtypes(include='object').columns
for col in string_columns:
    df[col] = df[col].apply(clean_text)
df.tail(60)
```

Level 1 Task 1

```python
#Top 3 cuisines in the datasets
df['Cuisines']=df['Cuisines'].str.split(",")
df=df.explode("Cuisines")
count_of_cuisine=df["Cuisines"].value_counts()
top_3_cuisine=count_of_cuisine.head(3)
print(f"Top 3 cuisines:{top_3_cuisine}")
#% of the top 3 cuisines in the datasets
total_restaurants=df["Restaurant ID"].nunique()
top_3_restaurants=(top_3_cuisine/total_restaurants)*100
print(f"% of top 3 cuisines:{top_3_restaurants}")
```

Level 1 Task 2

```python
#The city with highest number of restaurants
city_restaurants_counts=df["City"].value_counts()
city_with_most_restaurants=city_restaurants_counts.idxmax()
print(f"City with highest no of restaurants:{city_with_most_restaurants}")

#Average ratings of restaurants in each city
average_rating=df.groupby("City")["Aggregate rating"].mean()
print(f"Average ratings for restaurants in each city:{average_rating}")
city_with_highest_avg_rating=average_rating.idxmax()
print(f"City with the highest average ratings:{city_with_highest_avg_rating}")
```

Level 1 Task 3

```python
#Distribution of price ranges in bargraph
price_counts=df["Price range"].value_counts()
print(price_counts)
price_percentages=(price_counts/len(df))*100
print(price_percentages)
price_counts.plot(kind="bar",color=["green","blue","red","violet"])
plt.title("Distribution of price ranges among restaurants")
plt.xlabel("Price range")
plt.ylabel("No.of restaurants")
plt.show()
```

Level 1 Task 4

```python
#% of restaurants with online and offline delivery
total_restaurants = len(df)
online_delivery_count= df[df['Has Online delivery'] == 'Yes'].shape[0]
print(f"No of restaurants with online delivery:{online_delivery_count}")

percentage_online_delivery = (online_delivery_count / total_restaurants) * 100
print(f"% of restaurants with online delivery:{percentage_online_delivery}")

average_rating_with_delivery = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
average_rating_without_delivery = df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()
print(f"Average ratings of restaurants with online delivery:{average_rating_with_delivery}")
print(f"Average ratings of restaurants without online delivery:{average_rating_without_delivery}")
```

Level 2 Task 1

```python
#Distribution of average ratings and average number of votes
rating_distribution = pd.cut(df['Aggregate rating'], bins=[0, 1, 2, 3, 4, 5])
most_common_range = rating_distribution.value_counts().idxmax()
average_votes = df['Votes'].mean()
print(f"Most common rating range: {most_common_range}")
print(f"Average number of votes: {average_votes:.2f}")
```

Level 2 Task 2

```python
#Highest average ratings
cuisine_counts = df['Cuisines'].value_counts()
avg_rating_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
print(f"Average rating by each cuisine:{avg_rating_by_cuisine}")

highest_rated_cuisines = avg_rating_by_cuisine.idxmax()
highest_avg_rating = avg_rating_by_cuisine.max()
print(f"Highest average ratings:{highest_avg_rating}")
```

Level 3 Task 2

```python
#Restaurants with highest and lowest votes
highest_votes = df.loc[df["Votes"].idxmax()]
lowest_votes = df.loc[df["Votes"].idxmin()]

print("Restaurant with Highest Votes:")
print(highest_votes)

print("\nRestaurant with Lowest Votes:")
print(lowest_votes)

# Correlation between votes and ratings
correlation = df["Votes"].corr(df["Aggregate rating"])
print(f"\nCorrelation between Votes and Rating: {correlation:.2f}")
```

Level 3 Task 3

```python
numeric_columns = ['Restaurant ID', 'Country Code', 'Longitude', 'Latitude', 
                   'Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Standardize column names
df.columns = [col.strip().replace(' ', '_').lower() for col in df.columns]

# Handling missing values
df.fillna('', inplace=True)

# Checking unique values in columns
print("Price Range:", df['price_range'].unique())
print("Has Online Delivery:", df['has_online_delivery'].unique())
print("Has Table Booking:", df['has_table_booking'].unique())

#Group and analyze data by price range
delivery_booking_analysis = df.groupby('price_range').agg(
    online_delivery_percentage=('has_online_delivery', lambda x: (x == 'Yes').mean() * 100),
    table_booking_percentage=('has_table_booking', lambda x: (x == 'Yes').mean() * 100),
    count=('price_range', 'size')  # Total count of restaurants in each price range
).reset_index()

print(delivery_booking_analysis)

# Visualization
plt.figure(figsize=(10, 6))
plt.bar(delivery_booking_analysis['price_range'] - 0.2, delivery_booking_analysis['online_delivery_percentage'], width=0.4, label='Online Delivery')
plt.bar(delivery_booking_analysis['price_range'] + 0.2, delivery_booking_analysis['table_booking_percentage'], width=0.4, label='Table Booking')

plt.xlabel('Price Range')
plt.ylabel('Percentage (%)')
plt.title('Percentage of Online Delivery and Table Booking Availability by Price Range')
plt.legend()
plt.xticks(delivery_booking_analysis['price_range'])

plt.show()
```
